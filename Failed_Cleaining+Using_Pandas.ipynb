```python
csv_database = create_engine('sqlite:///csv_database.db')
```

```python
chunksize = 50000
i = 0
j = 1
for df in pd.read_csv("3gb_test.csv", chunksize=chunksize, iterator=True,delimiter="\t",encoding='latin1',dtype=str):
      df = df.rename(columns={c: c.replace(' ', '') for c in df.columns}) 
      df.index += j
      i+=1
      df.to_sql('table', csv_database, if_exists='append')
      j = df.index[-1] + 1
```
